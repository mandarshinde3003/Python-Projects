<h4> SMART AUTO FARE ESTIMATOR SYSTEM

```python
# SMART AUTO-RICKSHAW FARE ESTIMATOR & TRIP LOGGER
# Uses class attribute, instance methods, class method, static method, constructor & loops

class AutoFare:

    # class attribute – same for all autos
    company_name = "Mumbai Auto Fare Services"

    def __init__(self):
        self.base_fare = 25        # starting meter
        self.per_km_rate = 15      # ₹ per km
        self.waiting_charge = 2    # ₹ per minute
        self.trip_history = []     # stores past trips

    # static method – login system (same style as your notes)
    @staticmethod
    def login():
        name = input("Enter username: ")
        pwd = input("Enter password: ")
        if pwd == "Grantaccess":
            return "Login Successful"
        else:
            return "Login Failed"

    # class method – show organization name
    @classmethod
    def show_company(cls):
        return f"Fare system managed by: {cls.company_name}"

    # instance method – calculates total fare
    def calculate_fare(self, distance, waiting):
        total = self.base_fare + (distance * self.per_km_rate) + (waiting * self.waiting_charge)

        # add trip to history
        self.trip_history.append({
            "distance": distance,
            "waiting": waiting,
            "fare": total
        })

        return total

    # instance method – show trip history
    def show_history(self):
        if not self.trip_history:
            return "No trips recorded yet!"

        print("Trip History")
        for i, trip in enumerate(self.trip_history, start=1):
            print(f"Trip {i}: {trip['distance']} km, {trip['waiting']} min waiting → ₹{trip['fare']}")
        return ""

## Here's the main code 
print("=====================================")
print("   SMART AUTO FARE ESTIMATOR SYSTEM   ")
print("=====================================")

print(AutoFare.login())  # static method

auto = AutoFare()        # object created

print("\n" + AutoFare.show_company())  # class method

while True:
    print("\n1. Calculate Fare")
    print("2. Show Trip History")
    print("3. Exit")

    choice = input("Enter choice (1-3): ")

    if choice == "1":
        dist = float(input("Enter distance travelled (km): "))
        wait = int(input("Enter waiting time (minutes): "))
        total_fare = auto.calculate_fare(dist, wait)
        print(f"\nTotal Fare: ₹{total_fare}")

    elif choice == "2":
        auto.show_history()

    elif choice == "3":
        print("Thank you for using the Auto Fare Estimator!")
        break

    else:
        print("Invalid choice! Try again.")
```
